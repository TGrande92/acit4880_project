# K-Means Clustering

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
dataset = pd.read_csv('Mall_Customers.csv')
X = dataset.iloc[:, [3, 4]].values
```

## Using the elbow method to find the optimal number of clusters

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

## Training the K-Means model on the dataset

```python
kmeans = KMeans(n_clusters = 5, init = 'k-means++', random_state = 42)
y_kmeans = kmeans.fit_predict(X)
```

## Visualising the clusters

```python
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s = 100, c = 'green', label = 'Cluster 3')
plt.scatter(X[y_kmeans == 3, 0], X[y_kmeans == 3, 1], s = 100, c = 'cyan', label = 'Cluster 4')
plt.scatter(X[y_kmeans == 4, 0], X[y_kmeans == 4, 1], s = 100, c = 'magenta', label = 'Cluster 5')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'yellow', label = 'Centroids')
plt.title('Clusters of customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```

# Q2. 
### Start with 400 pairs of random values between 0 and 1. 
### Plot the data in a scatter plot
### Add extra data points in 4 different regions to the data set (400 new data points, 100 for each region) to artificially create 4 clusters in the entire data set. You can even use more points if you like.
### Apply K-Mean Clustering on the data. You should have 4 clusters (4 centers).
### Plot the center of each cluster in yellow and data points of each cluster in different colors.


```python
from sklearn.preprocessing import StandardScaler

customer_df = pd.read_csv('Social_Network_Ads.csv')

# Remove purchased
customer_df.drop(['Purchased'], axis=1, inplace=True)

wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(customer_df)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()

scaler = StandardScaler()
scaler.fit(customer_df)
scaled_features = scaler.transform(customer_df)
df_feat = pd.DataFrame(scaled_features, columns=customer_df.columns)

kmeans = KMeans(n_clusters = 3, init = 'k-means++', random_state = 42)
y_kmeans = kmeans.fit_predict(df_feat)

plt.scatter(df_feat.loc[y_kmeans == 0, 'Age'], df_feat.loc[y_kmeans == 0, 'EstimatedSalary'], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(df_feat.loc[y_kmeans == 1, 'Age'], df_feat.loc[y_kmeans == 1, 'EstimatedSalary'], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(df_feat.loc[y_kmeans == 2, 'Age'], df_feat.loc[y_kmeans == 2, 'EstimatedSalary'], s = 100, c = 'green', label = 'Cluster 3')

plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'black', label = 'Centroids')

plt.title('Clusters of customers')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()

```

Question 2

```python
data = np.random.rand(400, 2)

regions = {
    "region1": {"center": [0.2, 0.2], "variance": 0.1},
    "region2": {"center": [0.8, 0.2], "variance": 0.1},
    "region3": {"center": [0.2, 0.8], "variance": 0.1},
    "region4": {"center": [0.8, 0.8], "variance": 0.1}
}

for region in regions.values():
    new_data = np.random.normal(loc=region["center"], scale=region["variance"], size=(100, 2))
    data = np.vstack([data, new_data])


kmeans = KMeans(n_clusters=4, random_state=0)
kmeans.fit(data)
labels = kmeans.labels_
centers = kmeans.cluster_centers_

colors = ['red', 'blue', 'green', 'purple']
plt.figure(figsize=(8, 6))
plt.scatter(data[labels==0, 0], data[labels==0, 1], s=50, c=colors[0], label='Cluster 1')
plt.scatter(data[labels==1, 0], data[labels==1, 1], s=50, c=colors[1], label='Cluster 2')
plt.scatter(data[labels==2, 0], data[labels==2, 1], s=50, c=colors[2], label='Cluster 3')
plt.scatter(data[labels==3, 0], data[labels==3, 1], s=50, c=colors[3], label='Cluster 4')
plt.scatter(centers[:, 0], centers[:, 1], s=200, c='yellow', label='Centroids', edgecolors='black')

plt.title("K-Means Clustering with 4 Clusters")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```
